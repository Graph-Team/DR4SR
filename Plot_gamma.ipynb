# Toy

```python


import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import LinearLocator
# generate data
matplotlib.rcParams['xtick.labelsize'] = 16
matplotlib.rcParams['ytick.labelsize'] = 16

x = ['0', '0.1', '0.3', '0.5', '1']

recall = []
ndcg = []


# Start with subplots
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.yaxis.set_major_locator(LinearLocator(5))
ax1.set_ylim(0.094, 0.1080)

ax2.yaxis.set_major_locator(LinearLocator(5))
ax2.set_ylim(0.045, 0.053)

ax1.set_ylabel('Recall@20', fontdict={'size': 16})
ax2.set_ylabel('NDCG@20', fontdict={'size': 16})

lns1 = ax1.plot(x, recall, label='Recall@20', color=color[0], lw=2, marker='o', ms=10)
lns2 = ax2.plot(x, ndcg, label='NDCG@20', color=color[1], lw=2, marker='s', ms=10)

lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)
# plt.tick_params(labelsize=23)
plt.savefig('./figure/gamma_toy.pdf', format='pdf', bbox_inches = 'tight')
plt.show()
```

```python


import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import LinearLocator
# generate data
matplotlib.rcParams['xtick.labelsize'] = 16
matplotlib.rcParams['ytick.labelsize'] = 16

x = ['0', '0.1', '0.3', '0.5', '1']

recall = [0.0893, 0.0914, 0.0905, 0.0894, 0.0839]
ndcg = [0.0385, 0.0398, 0.0390, 0.0391, 0.0359]


# Start with subplots
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

ax1.yaxis.set_major_locator(LinearLocator(5))
ax1.set_ylim(0.081, 0.0940)

ax2.yaxis.set_major_locator(LinearLocator(5))
ax2.set_ylim(0.0340, 0.041)

ax1.set_ylabel('Recall@20', fontdict={'size': 16})
ax2.set_ylabel('NDCG@20', fontdict={'size': 16})

lns1 = ax1.plot(x, recall, label='Recall@20', color=color[0], lw=2, marker='o', ms=10)
lns2 = ax2.plot(x, ndcg, label='NDCG@20', color=color[1], lw=2, marker='s', ms=10)

lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)
# plt.tick_params(labelsize=23)
plt.savefig('./figure/gamma_beauty.pdf', format='pdf', bbox_inches = 'tight')
plt.show()
```
