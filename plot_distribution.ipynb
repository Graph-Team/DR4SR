```python
import os
import torch
import matplotlib.pyplot as plt
```

```python
dataset_name = 'book'
path = os.path.join('seq_len', dataset_name + '.seq_len')
data = torch.load(path, map_location='cpu')
```

```python
plt.hist(data, bins='auto')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'{dataset_name} distribution')
plt.show()
```
